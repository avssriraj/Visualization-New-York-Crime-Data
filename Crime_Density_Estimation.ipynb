```python
from shapely.geometry import LineString
import pandas as pd


j1 = {"x1": mean_lat1, "y1": mean_long1,
     "x2": data1["Latitude"], "y2": data1["Longitude"],}
dataf1 = pd.DataFrame(j1)

j2 = {"x1": mean_lat2, "y1": mean_long2,
     "x2": data2["Latitude"], "y2": data2["Longitude"],}
dataf2 = pd.DataFrame(j2)

j3 = {"x1": mean_lat3, "y1": mean_long3,
     "x2": data3["Latitude"], "y2": data3["Longitude"],}
dataf3 = pd.DataFrame(j3)
```

```python
def get_distance_1(k):
  lstr = LineString([(k.x1, k.y1,), (k.x2, k.y2) ])
  return lstr.length

def get_diff_x_1(l):
    v = (l.x1 - l.x2)**2
    return v

def get_diff_y_1(m):
    u = (m.y1 - m.y2) ** 2
    return u

def get_distance_2(k):
  lstr = LineString([(k.x1, k.y1,), (k.x2, k.y2) ])
  return lstr.length

def get_diff_x_2(l):
    v = (l.x1 - l.x2)**2
    return v

def get_diff_y_2(m):
    u = (m.y1 - m.y2) ** 2
    return u

def get_distance_3(k):
  lstr = LineString([(k.x1, k.y1,), (k.x2, k.y2) ])
  return lstr.length

def get_diff_x_3(l):
    v = (l.x1 - l.x2)**2
    return v

def get_diff_y_3(m):
    u = (m.y1 - m.y2) ** 2
    return u

dataf1["Dist"] = dataf1.apply(get_distance_1, axis=1)
dataf1["diff_sq_x"] = dataf1.apply(get_diff_x_1, axis = 1)
dataf1["diff_sq_y"] = dataf1.apply(get_diff_y_1, axis = 1)

dataf2["Dist"] = dataf2.apply(get_distance_2, axis=1)
dataf2["diff_sq_x"] = dataf2.apply(get_diff_x_2, axis = 1)
dataf2["diff_sq_y"] = dataf2.apply(get_diff_y_2, axis = 1)

dataf3["Dist"] = dataf3.apply(get_distance_3, axis=1)
dataf3["diff_sq_x"] = dataf3.apply(get_diff_x_3, axis = 1)
dataf3["diff_sq_y"] = dataf3.apply(get_diff_y_3, axis = 1)
```

```python
some_points_1 = []
some_points_2 = []
some_points_3 = []

some_points_1 = np.stack((npx1, npy1), axis=1)
some_points_2 = np.stack((npx2, npy2), axis=1)
some_points_3 = np.stack((npx3, npy3), axis=1)

print(some_points_1, some_points_2, some_points_3)
```

```python
distance_m1 = []
distance_m2 = []
distance_m3 = []


for i in all_points:
    for j in some_points_1:
        distance_m1.append(math.hypot(j[0]-i[0] , j[1]-i[1]))
        
for i in all_points:
    for j in some_points_2:
        distance_m2.append(math.hypot(j[0]-i[0] , j[1]-i[1]))
        
for i in all_points:
    for j in some_points_3:
        distance_m3.append(math.hypot(j[0]-i[0] , j[1]-i[1]))
        
```

```python
print(len(distance_m1), len(distance_m2), len(distance_m3))
```

```python
d_div_r1 = []
d_div_r2 = []
d_div_r3 = []

for d in distance_m1:
    d_div_r1.append(d/search_radius1)
for d in distance_m2:
    d_div_r2.append(d/search_radius2)
for d in distance_m3:
    d_div_r3.append(d/search_radius3)
```

```python
print(len(d_div_r1), len(d_div_r2), len(d_div_r3))
```

```python
sq_dr1 = []
sq_dr2 = []
sq_dr3 = []

for dr in d_div_r1:
    sq_dr1.append(math.pow((1 - math.pow(dr,2)), 2))
    
for dr in d_div_r2:
    sq_dr2.append(math.pow((1 - math.pow(dr,2)), 2))
    
for dr in d_div_r3:
    sq_dr3.append(math.pow((1 - math.pow(dr,2)), 2))

```

```python
print(len(sq_dr1), len(sq_dr2), len(sq_dr3))
```

```python
pi_sq_dr1 = []
pi_sq_dr2 = []
pi_sq_dr3 = []

for dr in sq_dr1:
    pi_sq_dr1.append( (3/math.pi)* dr)
    
for dr in sq_dr2:
    pi_sq_dr2.append( (3/math.pi)* dr)
    
for dr in sq_dr3:
    pi_sq_dr3.append( (3/math.pi)* dr)
    
```

```python
print(len(pi_sq_dr1), len(pi_sq_dr2), len(pi_sq_dr3))
```

```python
su_pi_sq_dr1 = np.add.reduceat(pi_sq_dr1, np.arange(0, len(pi_sq_dr1), 7))
su_pi_sq_dr2 = np.add.reduceat(pi_sq_dr2, np.arange(0, len(pi_sq_dr2), 2))
su_pi_sq_dr3 = np.add.reduceat(pi_sq_dr3, np.arange(0, len(pi_sq_dr3), 3))

print(len(su_pi_sq_dr1), len(su_pi_sq_dr2), len(su_pi_sq_dr3))
```

```python
supi_sq_dr1 = su_pi_sq_dr1.tolist()
supi_sq_dr2 = su_pi_sq_dr2.tolist()
supi_sq_dr3 = su_pi_sq_dr3.tolist()

print(len(supi_sq_dr1), len(supi_sq_dr3), len(supi_sq_dr3))
```

```python
density1 = []
density2 = []
density3 = []

for dr in supi_sq_dr1:
    density1.append( (1/search_radius1**2) * dr) 
    
for dr in supi_sq_dr2:
    density2.append( (1/search_radius2**2) * dr) 
    
for dr in supi_sq_dr3:
    density3.append( (1/search_radius3**2) * dr) 
    
print(len(density1), len(density3), len(density3))
```
